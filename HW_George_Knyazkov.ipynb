## Подключение бибилиотек

```python
# Работа с данными
import pandas as pd
import numpy as np

# Библиотеки для визуализации
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Начнем с чтения csv данных
df = pd.read_csv('sample_data/spotify-2023_fixed.csv')
```

```python
# Первые пять строк
df.head()
```

# **Columns** (Данные находящиеся в столбцах).

`1 - in_spotify_playlists:` Количество плейлистов Spotify, в которые включена песня.

`2 - in_spotify_charts:` Наличие и рейтинг песни в чартах Spotify.

`3 - streams:` Общее количество потоков на Spotify.

`4 - in_apple_playlists:` Количество плейлистов Apple Music, в которые включена песня.

`5 - in_apple_charts:` Наличие и рейтинг песни в чартах Apple Music.

`6 - in_deezer_playlists:` Количество плейлистов Deezer, в которые включена песня.

`7 - in_deezer_charts:` Наличие и рейтинг песни в чартах Deezer.

`8 - in_shazam_charts:` Наличие и рейтинг песни в чартах Shazam.

`9 - bpm:` Количество ударов в минуту, показатель темпа песни.

`10 - key:` Ключ композиции.

`11 - mode:` Режим песни (мажорный или минорный).

`12 - danceability_%:` Процентное соотношение, показывающее, насколько песня подходит для танцев.

`13 - valence_%:` Позитивность музыкального содержания песни.

`14 - energy_%:` Воспринимаемый уровень энергии песни.

`15 - acousticness_%:` Количество акустического звука в песне.

`16 - instrumentalness_%:` Количество инструментального содержания в песне.

`17 - liveness_%:` Наличие элементов живого исполнения.

`18 - speechiness_%:` Количество произнесенных слов в песне.

```python
title_dict = {'fontsize': 20, 'fontweight': 'bold', 'color': '#808080',
              'family': 'sans-serif'}
lable_dict = {'fontsize': 14, 'color': '#808080', 'family': 'sans-serif'}
```

# **Scatterplot** (Точечный график)

```python
# Установка размера графика для всех последующих графиков
sns.set(rc={'figure.figsize': (10, 6)})
sns.set_theme(style="whitegrid")
# Создание scatterplot с установленным размером
sns.scatterplot(data=df,
                x='in_spotify_charts',
                y='in_apple_charts',
                palette="crest",)

# Добавление заголовка к графику
plt.title("Spotify charts vs. Apple charts", fontdict=title_dict)

# Добавление подписей осей
plt.xlabel("Spotify charts", fontdict=lable_dict)
plt.ylabel("Apple charts", fontdict=lable_dict)

# Отображение графика
plt.show()
```

`in_spotify_charts:` Наличие и рейтинг песни в чартах Spotify.

`in_apple_charts:` Наличие и рейтинг песни в чартах Apple Music.


##Вывод по первому графику:

Из первого графика мы можем сделать вывод о том, что `рейтинг песен в чартах Apple Music` превалирует над `рейтингами песен в чартах Spotify`. Это говорит нам о том, что пользоваели `Apple Music` чаще обращаются к рейтингам песни, ориентируются на рейтинги песен соответственно. Это может говорить о большем колличестве пользователей `Apple Music`, но самом деле это не так, количество пользователей `Spotify` намного больше. Здесь скорее прослеживается разница **в подборках и рекомендациях для пользователей.** `Apple Music` больше ориентируется на рейтинги песен, а `Spotify` использует другие методы привлечения внимания аудитории к различным песням.

```python
# Установка размера графика для всех последующих графиков
sns.set(rc={'figure.figsize': (10, 10)})

# Создание scatterplot с установленным размером
sns.scatterplot(data=df,
                x='in_spotify_charts',
                y='in_apple_charts',
                hue='in_deezer_charts',
                palette="crest",)

# Добавление заголовка к графику
plt.title("Spotify vs. Apple vs. Deezer", fontdict=title_dict)

# Добавление подписей осей
plt.xlabel("Spotify charts", fontdict=lable_dict)
plt.ylabel("Apple charts", fontdict=lable_dict)

# Опционально добавление легенды
plt.legend(title='Deezer charts')

# Отображение графика
plt.show()
```

##Вывод по второму графику:

Добавив к рейтингам песен `в чартах Apple Music и Spotify` рейтинги песен в `Deezer`, мы наблюдаем, что в `Deezer` рейтингов песен еще меньше, чем в `Spotify и Apple Music`. Их кратно меньше, в несколько раз. **Основная масса рейтингов находится в первой четверти графика**, но в этом сегменте количество рейтингов не превышает и половины от максимасльных значений, а зачастую _равны нулю._ От середины и до максимальных значений рейтинги `Deezer` растянулись на оставшейся части графика. В этом сегменте _темные точки_, обозначающее максимальное количество рейтингов песен в `Deezer`, расположеные относительно `Apple Music и Spotify`, показывают эту весомую разницу, там где в `Deezer` значения равняются _50_, в `Apple Music` они больше _200_, а в `Spotify` приближаются к _120_. На примере одной из таких точек, мы еще раз убедились, что в `Deezer` **рейтингов кратно меньше, в несколько раз.** Вот такая кратная разница может уже свидетельствовать о количестве пользователей на `Deezer`, которых также кратно меньше, чем в `Spotify и Apple Music`.

```python
sns.set(rc={'figure.figsize': (10, 10)})

# Создание scatterplot с установленным размером
sns.scatterplot(data=df,
                x='in_spotify_charts',
                y='in_apple_charts',
                hue='in_deezer_charts',
                palette="flare",
                size='mode')
# Добавление заголовка к графику
plt.title("Spotify vs. Apple vs. Deezer + mode", fontdict=title_dict)

# Добавление подписей осей
plt.xlabel("Spotify charts", fontdict=lable_dict)
plt.ylabel("Apple charts", fontdict=lable_dict)

# Опционально добавление легенды
plt.legend(title='Deezer charts, mode')

# Отображение графика
plt.show()
```

## Гистограмма `histplot`.

`valence_%`: Позитивность музыкального содержания песни.

```python
sns.set(rc={'figure.figsize': (10, 6)})

sns.histplot(data=df, x="valence_%")

plt.title("Позитивность музыкального содержания песни", fontdict=title_dict)
plt.xlabel("Позитивность музыкального содержания песни",
              fontdict=lable_dict)
plt.ylabel("Соотношение", fontdict=lable_dict)

plt.show()
```

##Вывод по четвертому графику.

Здесь мне захотелось проанализировать `позитивность музыкального содержания в песнях`. По графику видно, что показатель _максимального соотношения_ у `песен` **со средней позитивностью (от 40 до 60)**. Основная масса песен находится в соотношении `60 и выше`. Из чего мы можем сделать вывод, что песни в чарте в большинстве своем имеют `позитивное содержание`, что и _повлияло на их попадание в топ_. Еще можно сказать, что люди в большинстве, так как мы говорим про топ чарт, предпочитают негативным или депрессивным песням более позитивные, это стандарнтая история, так как музыка, как отражает наше внутренне состояние, так и может его улучшить позитивными эмоциями, из чего можно сделать вывод, что люди нуждаются в позитивных эмоциях или находят отражение своего внутренного позитива в этих песнях.

##Новая переменная.

```python
df.loc[df['energy_%'] <= 50, 'energy_group'] = 'Спокойные'
df.loc[(df['energy_%'] > 50) & (df['energy_%'] <= 75),
       'energy_group'] = 'Средние'
df.loc[df['energy_%'] > 75, 'energy_group'] = 'Энергичные'
```

`energy_%:` Воспринимаемый уровень энергии песни.

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x="valence_%", hue="energy_group", palette="dark")

plt.title("Уровень энергии относительно позитивного содержания песни",
          fontdict=title_dict)
plt.xlabel("Позитивность музыкального содержания песни", fontdict=lable_dict)
plt.ylabel("Соотношение", fontdict=lable_dict)


plt.show()
```

##Вывод по пятому графику.

`Уровень энергии` в _максимальном соотношении позитивности_ `превалирует на среднем уровне`. Чем `выше позитивность` музыкального содержания в песне, тем `более усредненный уровень энергии`, хотя и есть исключение _в диапазоне позитивности около (60)_. Из чего мы можем сделать вывод, что **самые позитивные песни**, они же и **самые энергичные**, но даже у них _**уровень энергии не доходит до соотношения 30 и выше**_. В музыкальной индустрии не раз встречались песни с депрессивным или грустным содержанием, с очень веселой и энергичной музыкой, в данном чарте все довольно усреднено. Средний уровень позитивности сопряжен со средним уровнем энергии.

##Линейный график `relplot`

`danceability_%:` Процентное соотношение, показывающее, насколько песня подходит для танцев.

`instrumentalness_%:` Количество инструментального содержания в песне.

```python
plt.figure(figsize=(10, 6))  # Указываем ширину и высоту фигуры

# Для визуализации выбирается тип line
sns.relplot(data=df, x="danceability_%",
                    y="instrumentalness_%",
                    kind="line",
                    height=6,
                    aspect=1.67)

plt.title("Влияние танцевальности на Количество инструментального содержания",
                    fontdict=title_dict)
plt.xlabel("Насколько песня подходит для танцев", fontdict=lable_dict)
plt.ylabel("Количество инструментального содержания", fontdict=lable_dict)

plt.show()
```

##Вывод по шестому графику.

Довольно странная кореляция, которая дала не совсем ожидаемый результат. Казалось бы для _**танцев**_ `инструментальная музыка` не совсем подходит, мы привыкли немного к другому. И на первый взгляд все было логично. Наибольшие всплески `инструментального содержания` находятся в диапазоне, который говорит нам, что это `не совсем подходящее для танцев`, после 50 идут _небольшие угасающие скачки_, но `в самом конце`, где и есть _**максимальный диапазон, который подходит для танцев**_, мы видим подъем `инструментального содержания`. Из чего можно сделать вывод, что большиство песен `с инструментальным содержанием` _не подходит для танцев_, скорее всего это песни с большим количеством вокала или низким позитивным содержанием, но в конце графика мы видим, что `некоторое содержание инструментла`, все же _очень сгодится для `танцев`_.
